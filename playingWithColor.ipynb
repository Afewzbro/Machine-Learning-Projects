```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/sample_data/maxresdefault.jpg',0)
cv2_imshow(img)
plt.hist(img.ravel(),256,[0,256]);
plt.show()
src = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
dst = cv2.equalizeHist(src)
cv2_imshow(src)
cv2_imshow(dst)
print("...........................")
plt.hist(dst.ravel(),256,[0,256]); plt.show()
print("...........................")
plt.hist(dst.ravel(),256,[0,256]); plt.show()
```

```python
output2 = cv2.blur(img,(7,7))
cv2_imshow(output2)
print("...............................")
plt.show()
```
