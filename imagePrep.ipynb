```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/images.jpg')

cv2_imshow(img)
print(img.shape)
```

```python
from google.colab.patches import cv2_imshow
horizontalAppendedImg = np.hstack((img,img,img))
cv2_imshow(horizontalAppendedImg)
```

```python
verticalAppendedImg = np.vstack((img,img))
cv2_imshow(verticalAppendedImg)
```

```python
im2 = cv2.imread('/content/images.jpg',0)
cv2_imshow(im2)
```

```python
from google.colab.patches import cv2_imshow
image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(image)
```
